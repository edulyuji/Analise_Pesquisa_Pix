```python
import pandas as pd
from datetime import datetime
```

```python
Painelista = pd.read_csv('Dados/Painelista.csv', sep = ';')
Painelista_Pesquisa = pd.read_csv('Dados/Painelista_Pesquisa.csv', sep = ';')
Pesquisa = pd.read_csv('Dados/Pesquisa.csv', sep = ';')
Questao_DescResposta = pd.read_csv('Dados/Questao_DescResposta.csv', sep = ';')
Respostas = pd.read_csv('Dados/Respostas.csv', sep = ';')
```

```python
# 1) Faça a união dos 5 arquivos de dados e exporte uma única tabela contendo todas as respostas obtidas através dos 600 painelistas nas 2 pesquisas. 

#Definir o caminho para exportar o arquivo final
caminho_arquivo = 'Dados/dados.csv'

#Mesclar todas as colunas fornecidas
Dataframe_completo = pd.merge(Painelista, Painelista_Pesquisa, on='PainelistaID', how='right')
Dataframe_completo = pd.merge(Dataframe_completo, Pesquisa, on='PesquisaID', how='left')
Dataframe_completo = pd.merge(Dataframe_completo, Respostas, on=['CodCliente', 'CodJob'], how='inner')
Questao_DescResposta['Resposta'] = Questao_DescResposta['Resposta'].astype(str)
Dataframe_completo = pd.merge(Dataframe_completo, Questao_DescResposta, on=['CodQuestao','Resposta'], how='left')

# Exportar o dataframe para um arquivo CSV
Dataframe_completo.to_csv(caminho_arquivo, index=False, sep = ';', encoding='utf-8')
Dataframe_completo
```

```python
# A) Qual é a média e desvio padrão de idade dos 600 Painelista?

# Criar a coluna de Idade
Painelista_idade1 = Painelista.copy()
Painelista_idade1['Nascimento'] = pd.to_datetime(Painelista_idade1['Nascimento'], format='%m/%Y')
data_atual1 = datetime.now()
Painelista_idade1['idade'] = ((data_atual1 - Painelista_idade1['Nascimento']) / pd.Timedelta(days=365.25)).astype(int)


# Media
media_idade1 = Painelista_idade1['idade'].mean()
media_idade1 = round(media_idade1,2)

# Desvio Padrao
desvio_padrao1 = Painelista_idade1['idade'].std()
desvio_padrao1 = round(desvio_padrao1,2)

print( "A media de idade dos 600 Painelista é de:", media_idade1)
print( "O desvio padrão dos 600 Painelista é de:", desvio_padrao1)
```

```python
# B) Quantos % de quem possui filho(s) são do sexo Feminino?

# Excluir colunas desnecessarias
Painelista_feminino2 = Painelista.drop(columns=['Cidade', 'Estado','Nascimento'])

# Criacao do DF com os valores do Sexo Feminino
Painelista_feminino2 = Painelista_feminino2.query("Sexo == 'F'")

# Criar o DF das pesquisas realizadas
Painelista_pesquisas_realizadas2 = Painelista_Pesquisa.copy()

# Mesclar a tabela Painelista_feminino e Painelista_pesquisas_realizadas
Porcentagem_maes2 = pd.merge(Painelista_feminino2, Painelista_pesquisas_realizadas2, on='PainelistaID', how='inner')

# Filtrar dados de acordo com a pergunta desejada 
Respostas_filhos2 = Respostas.query("(DescQuestaoPOR == 'Você tem filhos?')")

# Mesclar a tabela Porcentagem_maes e Respostas_filhos
Porcentagem_maes2 = pd.merge(Porcentagem_maes2, Respostas_filhos2, on='CodCliente', how='inner')

# Obter a porcentagem
Porcentagem_maes2 = round(((Porcentagem_maes2.query("Resposta == '2'").shape[0] * 100)/Porcentagem_maes2.shape[0]))
print("A porcentagem de maes entrevistadas foi de:", Porcentagem_maes2,"%")
```

```python
# C) Quantos % dos Painelistas da região SUL possuem renda mensal até R$ 2.999?

# Exclui algumas colunas desnecessarias e Filtra apenas os estados da regiao SUL
Painelista_pessoas3 = Painelista.copy()
Painelista_pessoas3 = Painelista_pessoas3.drop(columns=['Cidade', 'Sexo', 'Nascimento'])
Painelista_pessoas3 = Painelista_pessoas3.query("Estado in ['SC', 'PR', 'RS']")

# Obter a listagem da pesquisa
Painelista_pesquisas_realizadas3 = Painelista_Pesquisa.copy()

# Mescla as pesquisas realizadas com as pessoas entrevistadas
Pessoas_Entrevistadas3 = pd.merge(Painelista_pessoas3, Painelista_pesquisas_realizadas3, on='PainelistaID', how='inner')

# fILTRA OS VALORES TOTAIS
Respostas_valores_total3 = Questao_DescResposta.query("DescOpcaoPOR in ['Até 720,00', 'R$ 721 a 1.100', 'R$ 1.101 a 1.999', 'R$ 2.000 a 2.999', 'R$ 3.000 a 3.999', 'R$ 4.000 a 4.999', 'R$ 5.000 a 6.999', 'R$ 7.000 a 9.999', 'R$ 10.000 a 14.999', 'R$ 15.000 a 20.000', 'Acima de 20.000']")

# Transforma a coluna Resposta para o formato de String
Respostas_valores_total3.loc[:, 'Resposta'] = Respostas_valores_total3['Resposta'].astype(str)

# Mescla as tabelas com valores TOTAIS e perguntas relacionadas
Respostas_valores_total3 = pd.merge(Respostas_valores_total3,Respostas,  on=['CodQuestao','Resposta'], how='inner')

# Mescla a tabela com os valores Totais com a tabela de pesquisa para obter os valores de PesquisaID
Respostas_valores_total3 = pd.merge(Pesquisa,Respostas_valores_total3,  on=['CodJob'], how='inner')

# Mescla as tabelas a tabela de pessoas entrevistadas com a tabela que contem todos os valores entrevistados
Respostas_valores_total3 = pd.merge(Pessoas_Entrevistadas3,Respostas_valores_total3,  on=['CodCliente','PesquisaID'], how='inner')

# filtra os valores desejados
Respostas_valores_desejado3 = Respostas_valores_total3.query("DescOpcaoPOR in ['Até 720,00', 'R$ 721 a 1.100', 'R$ 1.101 a 1.999', 'R$ 2.000 a 2.999']")

# Calcular a porcentagem
resultado_porcentagem3 = round(((Respostas_valores_desejado3.shape[0] * 100)/Respostas_valores_total3.shape[0]),2)

# Imprimir resultado
print("A porcentagem de Painelistas da região SUL que possuem renda mensal até R$ 2.999 é de:", resultado_porcentagem3,"%")
```

```python
# D) Quantos % dos 600 Painelistas usam o Pix para cada uma das finalidades pesquisadas?

# Obter a listagem da pesquisa
Painelista_pesquisas_realizadas4 = Painelista_Pesquisa.copy()

# Obter as pessoas pesquisadas e Mesclar as pesquisas realizadas com as pessoas entrevistadas
Pessoas_entrevistadas4 = Painelista.copy() 
Pessoas_entrevistadas4= pd.merge(Pessoas_entrevistadas4, Painelista_pesquisas_realizadas4, on='PainelistaID', how='inner')

#Filtrar apenas a pergunta desejada
pergunta_desejada4 = Respostas.query("DescQuestaoPOR == 'Para quais finalidades você usa o PIX? '")

# Mesclar a tabela da pergunta desejada com os resultados desejados
Perguntas_Respostas4 =  pd.merge(pergunta_desejada4, Questao_DescResposta, on='CodQuestao', how='inner')

# Mesclar a tabela da  com os resultados desejado
Perguntas_Respostas4 =  pd.merge(Perguntas_Respostas4, Pesquisa, on='CodJob', how='inner')


Perguntas_Respostas4 =  pd.merge(Perguntas_Respostas4, Pessoas_entrevistadas4, on='CodCliente', how='inner')

# Verificar se a coluna RESPOSTA_Y contem algum valor da coluna RESPOSTA_X
valores = []
for index, row in Perguntas_Respostas4.iterrows():
    if any(char in str(row['Resposta_y']) for char in str(row['Resposta_x'])):
        valores.append(index)

#Criar um dataframe com apenas dados necessarios
Perguntas_Respostas4 = Perguntas_Respostas4.iloc[valores]

# Calcular a porcentagem
frequencia = round((Perguntas_Respostas4['DescOpcaoPOR'].value_counts(normalize=True) * 100), 2).astype(str) + " %"

frequencia
```

```python
# E) Quantos % dos Painelistas que usam o Pix para [Transferir $$ para amigos e familiares] OU [Pagar compras em lojas online] são do sexo Masculino?

# Obter a listagem da pesquisa
Painelista_pesquisas_realizadas5 = Painelista_Pesquisa.copy()

# Obter as pessoas pesquisadas e Mesclar as pesquisas realizadas com as pessoas entrevistadas
Pessoas_entrevistadas5 = Painelista.copy() 
Pessoas_entrevistadas5 = pd.merge(Pessoas_entrevistadas5, Painelista_pesquisas_realizadas5, on='PainelistaID', how='inner')

#Filtrar apenas a pergunta desejada
pergunta_desejada5 = Respostas.query("DescQuestaoPOR == 'Para quais finalidades você usa o PIX? '")

# Mesclar a tabela da pergunta desejada com os resultados desejados
Perguntas_Respostas5 =  pd.merge(pergunta_desejada5, Questao_DescResposta, on='CodQuestao', how='inner')

# Mesclar a tabela da  com os resultados desejados
Perguntas_Respostas5 =  pd.merge(Perguntas_Respostas5, Pesquisa, on='CodJob', how='inner')

# filtra os valores desejados
Perguntas_Respostas5 = Perguntas_Respostas5.query("DescOpcaoPOR in ['Transferir $$ para amigos e familiares ', 'Pagar compras em lojas online ']")
Perguntas_Respostas5 =  pd.merge(Perguntas_Respostas5, Pessoas_entrevistadas5, on='CodCliente', how='inner')

# Excluir colunas desnecessarias
Perguntas_Respostas5 = Perguntas_Respostas5.drop(columns=['PainelistaID', 'PesquisaID_x','DescQuestaoPOR', 'Resposta_x', 'Resposta_y','Nome', 'Cidade', 'Estado', 'Nascimento','PesquisaID_y', 'Data'])
Perguntas_Respostas5

# Calcular a porcentagem
Porcentagem_desejada5 = round(((Perguntas_Respostas5.query("Sexo == 'M'").shape[0] * 100)/Perguntas_Respostas5.shape[0]),2)
print("A porcentagem de Painelistas que usam o Pix para [Transferir $$ para amigos e familiares] OU [Pagar compras em lojas online] e são do sexo Masculino é de:", Porcentagem_desejada5,"%")
```
